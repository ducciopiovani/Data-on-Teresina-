```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
d = pd.read_csv('Double_Soersen_Index.csv',header=None)
s = pd.read_csv('Single_Soersen_Index.csv',header=None)
```

```python
from matplotlib import rcParams
```

```python
rcParams["font.size"]= 20
```

```python
s.set_index(0).max()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize=(15,7))
s.plot(x=0,y=1,ax=ax1,label='Single Constrained',linewidth=4,c='r')
d.plot(x=0,y=1,ax=ax1,label='Double Constrained',linewidth=4,c='orange')
ax1.set_xlabel(r'$\beta$',fontsize=20)
ax1.set_ylabel('Soersen Index',fontsize=20)
old_labels = [l for l in ax1.get_yticks()]
labels =  [format(l,'0.2f') for l in old_labels]
ax1.set_yticklabels(labels)
ax1.set_title('Gravity Models', fontsize=20)
ax1.annotate('(a)', xy=(0.18, 0.38),fontsize=25)


e = pd.read_csv('Extended_Index.csv',header=None)
e.plot(x=0,y=1,ax=ax2,linewidth=4,legend=False)
ax2.annotate('(b)', xy=(1.55, 0.335),fontsize=25)
ax2.set_xlabel(r'$\alpha$',fontsize=20)
ax2.set_ylabel('Soersen Index',fontsize=20)

old_labels = [l for l in ax2.get_yticks()]
labels =  [format(l,'0.2f') for l in old_labels]
ax2.set_yticklabels(labels)
ax2.set_xlim(0,1.8)
ax2.set_title('Extended Radiation Model', fontsize=20)
plt.tight_layout()
plt.savefig("calibration.eps")
```
