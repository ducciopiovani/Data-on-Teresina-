```python
I = pd.read_csv('../SoersenIndexData/Extended_Index.csv',header=None)

import matplotlib as mpl
label_size = 8

mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20


fig, ax = plt.subplots(figsize=(8,5))
ax.set_title("Extended Radiation Calibration",fontsize=15)
I.plot(x=0,y=1,ax=ax,label='Extended-Radiation',linewidth=3)
ax.plot(np.repeat(0.145,4),np.linspace(0.439,0.445,4), c='b',linestyle='--',linewidth=2)
ax.set_xlabel(r"$\alpha$",fontsize=20)
ax.set_ylabel('Soersen Index',fontsize=20)
plt.legend(loc=2)
ax.annotate(r'$\alpha_{max} \rightarrow 0$', xy=(0.145, 0.444), xytext=(0.25, 0.35),fontsize=20)
plt.savefig('ExtendedCalibration.eps')
```

```python
Id = pd.read_csv('../SoersenIndexData/Double_Soersen_Index.csv',header=None)
Is = pd.read_csv('../SoersenIndexData/Single_Soersen_Index.csv',header=None)

import matplotlib as mpl

mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20

fig, ax = plt.subplots(figsize=(8,5))
ax.set_title("Gravity Models'Calibration",fontsize=15)

#plt.plot(Id[0],Id[1],label=)
Id.plot(x=0,y=1,ax=ax,label='Double-Constrained',linewidth=3,c='r')
Is.plot(x=0,y=1,ax=ax,label='Single-Constrained',linewidth=3,c='orange')

ax.plot(np.repeat(0.145,4),np.linspace(0.439,0.445,4), c='r',linestyle='--',linewidth=2)

ax.set_xlabel(r"$\beta$",fontsize=20)
ax.set_ylabel('Soersen Index',fontsize=20)

ax.annotate(r'$\beta^{double}_{max}=0.065$', xy=(0.065, 0.388), xytext=(0.075, 0.388),fontsize=20)
ax.annotate(r'$\beta^{single}_{max}=0.045$', xy=(0.015, 0.39), xytext=(0.010, 0.41),fontsize=20)

ax.plot(np.repeat(0.065,4), np.linspace(0.10,0.45,4), c='r',linestyle='--',linewidth=2)
ax.plot(np.repeat(0.045,4), np.linspace(0.10,0.45,4), c='orange',linestyle='--',linewidth=2)
ax.set_ylim(0.10,0.45)
plt.legend(loc=3)
plt.savefig('GravityCalibration.eps')
```
